```python
import numpy as np
from skimage.io import imread
from skimage.color import rgb2lab, lab2rgb
import matplotlib.pylab as plt
```

```python
im = imread('/content/sample_data/Tongue Pic. - Gulshan Rai.jpg')
im1 = rgb2lab(im)
```

Converting RGB image into grayscale by setting the
Lab space color channels to zero

```python
im = imread('/content/sample_data/Tongue Pic. - Gulshan Rai.jpg')
im1 = rgb2lab(im)
im1[...,1] = im1[...,2] = 0
im1 = lab2rgb(im1)
plt.figure(figsize=(20,10))
plt.subplot(121), plt.imshow(im), plt.axis('off'),
plt.title('Original image', size=20)
plt.subplot(122) 
plt.imshow(im1)
plt.axis('off')
plt.title('Grayscale image', size=20)
plt.show()
```

Changing the brightness of the image by varying the
luminosity channel

```python
im = imread('/content/sample_data/Tongue Pic. - Gulshan Rai.jpg')
im1 = lab2rgb(im1)
#im1 = rgb2lab(im)
im1[...,0] = im1[...,0] + 50
plt.subplot(121), plt.imshow(im), plt.axis('off'),
plt.title('Original image')
plt.subplot(122) 
plt.imshow(im1)
plt.axis('off')
plt.title('bright image')
plt.show()
```

```python
im1 = lab2rgb(im1)
im1 = rgb2lab(im)
im1[...,0] = im1[...,0] - 50
im1 = lab2rgb(im1)
plt.subplot(121), plt.imshow(im), plt.axis('off'),
plt.title('Original image')
plt.subplot(122) 
plt.imshow(im1)
plt.axis('off')
plt.title('dark image')
plt.show()
```

 obtain a more natural inverted image in the Lab space

```python
im1 = rgb2lab(im)
im1[...,0] = np.max(im1[...,0]) - im1[...,0]
im1 = lab2rgb(im1)
plt.subplot(121), plt.imshow(im), plt.axis('off'),
plt.title('Original image')
plt.subplot(122) 
plt.imshow(im1)
plt.axis('off')
plt.title('natural inverted image')
plt.show()
```
